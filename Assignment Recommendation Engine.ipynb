```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine , correlation
from sklearn.metrics.pairwise import cosine_similarity
```

```python
book_df = pd.read_csv(r"C:\Users\user\Downloads\book (4).csv" , encoding= ('ISO-8859-1') )
```

```python
book_df
```

```python
## rename the features

data = book_df.rename(columns = {'User.ID':'User_id' , 'Book.Title':'Book_title' , 'Book.Rating':'rating' , 'Unnamed: 0':'Unnamed'})
```

```python
data.head()
```

```python
data.sort_values('User_id')
```

```python
user_item = data.pivot_table(index = 'User_id', columns = 'Book_title' , values = 'rating').reset_index(drop = True)
```

```python
user_item
```

```python
user_item.fillna(0 , inplace = True)
```

```python
user_item
```

```python
## compute the cosine similarity

user_similarity = cosine_similarity(user_item.values)
```
