```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score , classification_report
```

```python
data = pd.read_csv(r"C:\Users\user\Downloads\SalaryData_Test.csv")
```

```python
x = data.drop(['Salary'] , axis = 1)
y = data['Salary']
```

```python
le = LabelEncoder()

x_encoded = x.apply(le.fit_transform)
```

```python
x_encoded
```

```python
y.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_encoded , y , test_size = 0.20 , random_state = 42)
```

```python
x_train.shape , x_test.shape
```

```python
## create a model

nb_classifier = GaussianNB()
nb_classifier.fit(x_train , y_train)
```

```python
## predict the value , use trained classifier to make the prediction on the test data sets

y_pred = nb_classifier.predict(x_test)
```

```python
y_pred
```

```python
## find out the accuracy of the model performace metrics


accuracy = accuracy_score(y_pred , y_test)
```

```python
accuracy
```

```python
## classification report

classification = classification_report(y_pred , y_test)
```
