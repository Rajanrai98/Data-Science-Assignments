```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score , classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(r"C:\Users\user\Downloads\SalaryData_Train.csv")
```

```python
data
```

```python
data.info()
```

```python
data.head()
```

```python
x = data.drop(['Salary'] , axis = 1)
y = data['Salary']
```

```python
le = LabelEncoder()

x_encoded = x.apply(le.fit_transform)
```

```python
x_encoded
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_encoded  , y , test_size = 0.20 , random_state = 42)
```

```python
## create a naive bayes classifier

nb_classifier = GaussianNB()

nb_classifier.fit(x_train , y_train)
```

```python
## predicted the value , use the trained classifier to make prediction on test data sets

y_pred = nb_classifier.predict(x_test)
```

```python
y_pred
```

```python
## calculate the accuracy

accuracy = accuracy_score(y_pred , y_test)
```

```python
accuracy
```

```python
## Generate the classification report

classification_rp = classification_report(y_pred , y_test)
```
