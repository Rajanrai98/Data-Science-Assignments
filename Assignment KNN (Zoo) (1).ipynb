```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report , accuracy_score
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(r"C:\Users\user\Downloads\Zoo.csv")
```

```python
data
```

```python
data.head()
```

```python
## split the data into the dependent variable and independent variable
```

```python
x = data.drop(['animal name' , 'type'] , axis =1)
y = data['type']
```

```python

y
```

```python
x.head()
```

```python
y = pd.DataFrame(y)
```

```python
y.head()
```

```python
## split the data for training and testing data sets

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 ,random_state = 42)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn_classifier
```

```python
## train the model 

knn_classifier.fit(x_train , y_train)
```

```python
knn_pred = knn_classifier.predict(x_test)
```

```python
knn_pred
```

```python
accuracy = accuracy_score(y_test , knn_pred)
```

```python
accuracy
```

```python
print(classification_report(y_test , knn_pred))
```

```python
## hyperparameter tuning 

param_grid = {'n_neighbors':[3,5,7,9],
             'weights':['uniform','distance'],
             'p':[1,2]}
```

```python
grid_search = GridSearchCV(knn_classifier , param_grid , cv = 5)
grid_search.fit(x_train , y_train)
```

```python
best_params = grid_search.best_params_
```

```python
best_params
```

```python
best_model = grid_search.best_estimator_

y_pred = best_model.predict(x_test)
```

```python
y_pred
```

```python
accuracy_sc = accuracy_score(y_test , y_pred)
```

```python
accuracy_sc
```

```python
print(classification_report(y_test , y_pred))
```
